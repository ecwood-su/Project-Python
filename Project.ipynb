```python
basketball = pd.read_csv('FourFactors.csv')
basketball
```

```python
Team_List = basketball['Team'].tolist()
FourFactor_List = basketball['FourFactors'].tolist()
Team_List
```

```python
FourFactor_df = pd.DataFrame( { 'Team' : Team_List, 'FourFactors' : FourFactor_List })
FourFactor_df
```

```python
title = '\033[1m' + 'Welcome to our College Basketball Matchup Predictor!'
print(title)

from IPython.core.display import HTML
while True : 
    print("Type 'quit' to exit: ")
    user_team1 = input('Enter a team: ')
    if user_team1.lower() == 'quit':
        break
    else:
        first = FourFactor_df.loc[(FourFactor_df.Team)==user_team1]
        user_team2 = input('Enter a team: ')
        second = FourFactor_df.loc[(FourFactor_df.Team)==user_team2]
        first_df = pd.DataFrame(first)
        second_df = pd.DataFrame(second)
        combined_list = [first_df, second_df]
        combined_df = pd.concat(combined_list)
        display(HTML(combined_df.to_html()))
        user_FourFactor_list = combined_df['FourFactors'].tolist()
        user_Team_list = combined_df['Team'].tolist()
        if user_FourFactor_list[0] > user_FourFactor_list[1]:
            print('We project', user_team1, 'to win the game')
            combined_df.plot.bar(x= 'Team', y= 'FourFactors')
            
        elif user_FourFactor_list[0] < user_FourFactor_list[1]:
            print('We project', user_team2, 'to win the game')
            combined_df.plot.bar(x= 'Team', y= 'FourFactors')

            
            
```
